# SOAP request on CLS Argos
Based on https://www.geeksforgeeks.org/making-soap-api-calls-using-python/

```python
with open('soap_credentials.txt') as fid:
    lines = fid.readlines()
username, password = lines[0].split()
```

```python
import zeep
  
# set the WSDL URL
wsdl_url = "http://ws-argos.cls.fr/argosDws/services/DixService?wsdl"
  
# set method URL
method_url = "http://www.w3.org/2003/05/soap-envelope"
  
# set service URL
service_url = "http://service.dataxmldistribution.argos.cls.fr/types"
  
# create the header element
header = zeep.xsd.Element(
    "Header",
    zeep.xsd.ComplexType(
        [
            zeep.xsd.Element(
                "{http://www.w3.org/2005/08/addressing}Action", zeep.xsd.String()
            ),
            zeep.xsd.Element(
                "{http://www.w3.org/2005/08/addressing}To", zeep.xsd.String()
            ),
        ]
    ),
)
# set the header value from header element
header_value = header(Action=method_url, To=service_url)
  
# initialize zeep client
client = zeep.Client(wsdl=wsdl_url)
```

```python
header_value
```

```python
result = client.service.getCsv(username=username,password=password,platformId='224110',nbDaysFromNow=10,)
```
